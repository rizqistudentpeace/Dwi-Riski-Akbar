## Penjelasan Statistika Parametrik dan Non-Parametrik

### Statistika Parametrik

**Definisi:**
Statistika parametrik adalah metode analisis statistik yang mengasumsikan bahwa data berasal dari populasi dengan distribusi tertentu (seringkali distribusi normal) dan memenuhi asumsi tertentu tentang parameter populasi (misalnya, varians homogen).

**Asumsi Kunci:**
1.  **Distribusi Normal:** Data berdistribusi normal atau mendekati normal.
2.  **Varians Homogen:** Varians antar kelompok yang dibandingkan adalah sama (homogen).
3.  **Skala Interval/Rasio:** Data diukur pada skala interval atau rasio.
4.  **Independensi:** Observasi dalam sampel bersifat independen.

**Keuntungan:**
*   Lebih kuat (powerful) dalam mendeteksi efek atau perbedaan jika asumsi terpenuhi.
*   Memberikan estimasi parameter populasi yang lebih presisi.

**Kerugian:**
*   Tidak cocok jika asumsi tidak terpenuhi, dapat menghasilkan kesimpulan yang salah.
*   Sensitif terhadap *outlier*.

**Contoh Metode:**
*   Uji-t (Independent/Paired Sample t-test)
*   ANOVA (Analysis of Variance)
*   Korelasi Pearson
*   Regresi Linier

**Contoh Kasus:**
Seorang peneliti ingin membandingkan rata-rata nilai ujian matematika antara dua kelompok siswa: satu kelompok diajar dengan metode A dan kelompok lain dengan metode B. Jika diasumsikan nilai ujian berdistribusi normal dan varians kedua kelompok homogen, peneliti dapat menggunakan **Uji-t Independen** untuk menentukan apakah ada perbedaan signifikan pada rata-rata nilai ujian kedua kelompok.

---

### Statistika Non-Parametrik

**Definisi:**
Statistika non-parametrik adalah metode analisis statistik yang tidak mengasumsikan distribusi data tertentu dari populasi. Metode ini lebih fleksibel dan digunakan ketika asumsi parametrik (terutama normalitas) tidak terpenuhi, atau ketika data diukur pada skala nominal atau ordinal.

**Asumsi Kunci:**
*   Tidak ada asumsi tentang bentuk distribusi data (distribusi bisa apa saja).
*   Biasanya digunakan untuk data berskala nominal atau ordinal, atau data interval/rasio yang tidak memenuhi asumsi normalitas.
*   Mungkin memiliki asumsi lain yang lebih longgar (misalnya, independensi).

**Keuntungan:**
*   Dapat digunakan untuk berbagai jenis data, termasuk data kualitatif dan data yang tidak berdistribusi normal.
*   Lebih tahan terhadap *outlier*.
*   Lebih mudah dipahami dan diinterpretasikan untuk beberapa metode.

**Kerugian:**
*   Kurang kuat (less powerful) dibandingkan uji parametrik jika asumsi parametrik terpenuhi.
*   Kurang efisien dalam menggunakan informasi data, terutama pada data interval/rasio.
*   Sulit untuk mengestimasi parameter populasi.

**Contoh Metode:**
*   Uji Wilcoxon Rank-Sum (ekuivalen non-parametrik dari Uji-t independen)
*   Uji Mann-Whitney U (mirip dengan Wilcoxon Rank-Sum)
*   Uji Kruskal-Wallis (ekuivalen non-parametrik dari ANOVA)
*   Korelasi Spearman
*   Uji Chi-Square

**Contoh Kasus:**
Seorang peneliti ingin membandingkan efektivitas dua obat baru (Obat X dan Obat Y) dalam mengurangi nyeri. Pasien diminta untuk menilai tingkat nyeri mereka pada skala ordinal (1 = sangat ringan, 2 = ringan, 3 = sedang, 4 = parah, 5 = sangat parah) setelah mengonsumsi obat. Karena data tingkat nyeri adalah ordinal dan mungkin tidak berdistribusi normal, peneliti dapat menggunakan **Uji Mann-Whitney U** untuk membandingkan peringkat tingkat nyeri antara kelompok yang menerima Obat X dan kelompok yang menerima Obat Y.

Berikan argumen anda mengenai “hasil penelitian yang valid pasti reliable dan obyektif, tapi apakah hasil penelitian yang reliable dan obyektif juga pasti valid”? Uraikan menurut perspektif anda?

## Argumen mengenai Validitas, Reliabilitas, dan Objektivitas dalam Penelitian

### Pendahuluan
Dalam metodologi penelitian, konsep validitas, reliabilitas, dan objektivitas adalah pilar penting untuk menilai kualitas dan kredibilitas suatu temuan. Meskipun sering dibahas bersama, ketiganya memiliki makna dan implikasi yang berbeda, serta hubungan hierarkis tertentu.

### 1. “Hasil Penelitian yang Valid Pasti Reliable dan Objektif”

**Argumen: YA, pernyataan ini benar.**

*   **Validitas (Validity):** Validitas mengacu pada sejauh mana suatu instrumen atau penelitian benar-benar mengukur apa yang seharusnya diukur. Jika sebuah penelitian dikatakan valid, itu berarti hasil yang diperoleh benar-benar merepresentasikan fenomena yang diteliti dan kesimpulan yang ditarik adalah akurat.

*   **Implikasi terhadap Reliabilitas:** Jika hasil penelitian valid, secara inheren ia harus reliable. Reliabilitas (reliability) adalah konsistensi atau stabilitas pengukuran. Jika suatu pengukuran valid, artinya ia mengukur dengan tepat apa yang dituju, maka pengukuran tersebut harus memberikan hasil yang konsisten jika diulang dalam kondisi yang sama. Tidak mungkin sebuah alat mengukur 'sesuatu' dengan benar (valid) tetapi hasilnya tidak konsisten (tidak reliable). Inkonsistensi akan mengindikasikan bahwa alat tersebut tidak benar-benar mengukur apa yang seharusnya, sehingga membatalkan validitasnya.

*   **Implikasi terhadap Objektivitas:** Validitas juga mensyaratkan objektivitas. Objektivitas berarti bahwa hasil penelitian tidak dipengaruhi oleh bias pribadi peneliti atau faktor-faktor subjektif lainnya. Jika sebuah penelitian valid, maka kesimpulan yang ditarik harus didasarkan pada bukti empiris yang kuat dan tidak memihak, sehingga dapat diterima oleh pihak lain dengan sudut pandang berbeda. Bias subjektif akan merusak akurasi pengukuran atau interpretasi, yang secara langsung mengancam validitas.

**Kesimpulan Bagian 1:** Validitas adalah konsep yang lebih tinggi yang mencakup reliabilitas dan objektivitas. Untuk menjadi valid, suatu pengukuran atau penelitian harus terlebih dahulu reliable (konsisten) dan objektif (bebas bias).

### 2. “Apakah Hasil Penelitian yang Reliable dan Objektif Juga Pasti Valid?”

**Argumen: TIDAK, pernyataan ini tidak selalu benar.**

*   **Reliabel tetapi Tidak Valid:** Sebuah pengukuran bisa sangat reliable (konsisten) tetapi tidak valid. Bayangkan sebuah timbangan yang selalu menunjukkan angka 5 kg setiap kali Anda menimbang sebuah benda, terlepas dari berat benda sebenarnya. Timbangan ini sangat reliable karena konsisten dalam memberikan hasil yang sama, tetapi sama sekali tidak valid karena tidak mengukur berat benda yang sebenarnya. Dalam penelitian, instrumen yang selalu memberikan hasil yang sama tetapi tidak relevan dengan konsep yang ingin diukur adalah reliable tetapi tidak valid.

*   **Objektif tetapi Tidak Valid:** Penelitian juga bisa bersifat objektif tetapi tidak valid. Objektivitas memastikan bahwa data dikumpulkan dan dianalisis tanpa bias, dan interpretasi dilakukan secara netral. Namun, objektivitas tidak secara otomatis menjamin bahwa apa yang diukur adalah *hal yang benar* atau *sesuai* dengan tujuan penelitian. Misalnya, seorang peneliti secara objektif mengumpulkan data tentang tinggi badan siswa, tetapi tujuan penelitiannya adalah mengukur kecerdasan. Data tinggi badan dikumpulkan secara objektif, namun tidak valid untuk mengukur kecerdasan.

**Contoh Kasus Gabungan:**
Seorang guru ingin mengukur kemampuan pemahaman siswa terhadap materi pelajaran. Guru tersebut menggunakan ujian pilihan ganda dengan soal-soal yang tidak relevan dengan materi yang diajarkan, tetapi memiliki kunci jawaban yang sangat jelas sehingga setiap siswa yang menjawab dengan cara tertentu akan selalu mendapatkan skor yang sama (misalnya, semua jawaban 'A' akan menghasilkan skor tinggi, terlepas dari pemahaman). Ujian ini mungkin memberikan hasil yang **reliable** (karena konsisten menghasilkan skor yang sama jika diberikan berulang kali kepada kelompok siswa yang sama) dan bahkan bisa dianggap **objektif** (karena penilaiannya baku, tidak ada bias personal). Namun, ujian tersebut sama sekali **tidak valid** karena tidak mengukur pemahaman siswa terhadap materi pelajaran yang sebenarnya.

### Kesimpulan Akhir

*   **Validitas adalah tujuan tertinggi** dalam penelitian. Sebuah penelitian yang valid adalah penelitian yang baik, yang secara otomatis akan reliable dan objektif.
*   **Reliabilitas dan objektivitas adalah kondisi yang diperlukan, tetapi tidak cukup,** untuk validitas. Mereka adalah prasyarat. Sebuah penelitian bisa saja konsisten dan bebas bias, tetapi jika tidak mengukur atau meneliti apa yang seharusnya, maka hasilnya tetap tidak valid.

Periksalah apakah data diameter batang tanaman berdistribusi normal atau
tidak dan lakukan uji hiptesis dgn α=0.05
**GUNAKAN SOFTWARE**

No Diameter(cm) No Diameter(cm) No Diameter(cm)
1 23 11 32 21 22

3 25 13 17 23 33
4 25 14 20 24 28
5 24 15 19 25 39
6 25 16 31 26 34
7 24 17 36 27 35
8 15 18 32 28 39

```python
#instal pandas
#distribusi normal atau tidak
#uji hipotesis α=0.05
import pandas as pd
data_tanaman = [23, 11, 32, 21, 22,
                24, 12, 12, 22, 23,
                25, 13, 17, 23, 33,
                25, 14, 20, 24, 28,
                24, 15, 19, 25, 39,
                25, 16, 31, 26, 34,
                24, 17, 36, 27, 35,
                15, 18, 32, 28, 39,
                34, 19, 27, 29, 29,
                33, 20, 29, 30, 11]
df = pd.DataFrame({'Nilai': data_tanaman})
print(df)
```

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Uji Normalitas Shapiro-Wilk
stat, p_value = stats.shapiro(df['Nilai'])

print(f"Statistik Uji Shapiro-Wilk: {stat:.3f}")
print(f"P-value: {p_value:.3f}")

alfa = 0.05
print(f"\u03B1 (tingkat signifikansi): {alfa}")

# Interpretasi hasil
if p_value > alfa:
    print("Keputusan: Gagal menolak H0. Data berdistribusi normal.")
else:
    print("Keputusan: Menolak H0. Data tidak berdistribusi normal.")

# Visualisasi histogram untuk inspeksi visual
plt.figure(figsize=(8, 6))
sns.histplot(df['Nilai'], kde=True, bins=5)
plt.title('Distribusi Diameter Batang Tanaman')
plt.xlabel('Diameter (cm)')
plt.ylabel('Frekuensi')
plt.show()
```

**Hipotesis:**
*   **H0 (Hipotesis Nol):** Data berdistribusi normal.
*   **Ha (Hipotesis Alternatif):** Data tidak berdistribusi normal.

**Penjelasan:**
*   Jika nilai p-value > \(\alpha\) (tingkat signifikansi, dalam kasus ini 0.05), kita gagal menolak H0. Ini berarti data dianggap berdistribusi normal.
*   Jika nilai p-value \(\le\) \(\alpha\), kita menolak H0. Ini berarti data tidak dianggap berdistribusi normal.

Kesimpulannya adalah data berdistribusi normal tidak pada tingkat signifikansi 0.05.

## Soal No-6
Efektifitas Tiga metode mengajar yaitu Diskusi, Problem Solving, dan Tugas terlihat
dari nilai matematika terhadap 3 kelompok siswa selama 3 bulan. Data hasil belajar
di sajikan sebagai berikut (Software Pyhton):

```python
import pandas as pd
import scipy.stats as stats

# Data hasil belajar untuk setiap metode
diskusi = [7, 7, 8, 9, 6, 7, 8, 8, 8, 8]
problem_solving = [6, 6, 6, 7, 7, 7, 8, 9, 8, 8]
tugas = [9, 9, 9, 9, 8, 8, 7, 8, 9, 9]

# Buat DataFrame
df_metode = pd.DataFrame({
    'Metode': ['Diskusi']*len(diskusi) + ['Problem Solving']*len(problem_solving) + ['Tugas']*len(tugas),
    'Nilai': diskusi + problem_solving + tugas
})

print("Data Hasil Belajar:")
print(df_metode.head())
print("\nDeskripsi Data:")
print(df_metode.groupby('Metode')['Nilai'].describe())
```

### Pengujian Homogenitas Varians (Levene's Test)

**Hipotesis:**
*   **H0 (Hipotesis Nol):** Varians dari ketiga kelompok metode mengajar adalah homogen (sama).
*   **Ha (Hipotesis Alternatif):** Varians dari setidaknya satu kelompok metode mengajar berbeda secara signifikan (tidak homogen).

**Tingkat Signifikansi (\(\alpha\)):** 0.05

```python
# Lakukan Levene's Test
# Menggunakan center='median' karena lebih robust terhadap non-normalitas
stat_levene, p_value_levene = stats.levene(diskusi, problem_solving, tugas, center='median')

print(f"Statistik Uji Levene: {stat_levene:.3f}")
print(f"P-value: {p_value_levene:.3f}")

alfa = 0.05
print(f"\u03B1 (tingkat signifikansi): {alfa}")

# Interpretasi hasil
if p_value_levene > alfa:
    print("Keputusan: Gagal menolak H0. Varians ketiga kelompok adalah homogen.")
else:
    print("Keputusan: Menolak H0. Varians ketiga kelompok tidak homogen.")

print("\nKesimpulan: Berdasarkan uji Levene dengan \u03B1 = 0.05, kami akan menyimpulkan apakah varians antar kelompok metode mengajar homogen atau tidak.")
```
