##Soal No-3
Diberikan data 60 mahasiswa dari prodi sains aktuaria universitas A berikut
Dari data tersebut carilah mean, modus, median, serta buatlah tabel distribusi
frekuensi?
##**Dataset:**

##Jawaban:

```python
import pandas as pd
data_mahasiswa = ([60,67, 63, 81, 64, 75, 64, 70, 73, 71,
                  83, 75, 56, 53, 73, 80, 71, 81, 89, 80,
                  57, 54, 60, 75, 76, 83, 65, 75, 81, 71,
                  74, 70, 64, 89, 64, 85, 51, 75, 55, 56,
                  89, 89, 74, 86, 74, 86, 61, 55, 56, 58,
                  77, 79, 51, 55, 65, 38, 77, 31, 33, 77])
df = pd.DataFrame({'Nilai': data_mahasiswa})
print(df)
print(df.describe())

# Menghitung Mean (Rata-rata)
mean_nilai = df['Nilai'].mean()
print(f"Nilai Mean: {mean_nilai}")
# Menghitung Median (Nilai Tengah)
median_nilai = df['Nilai'].median()
print(f"Nilai Median: {median_nilai}")
# Menghitung Modus (Nilai yang Sering Muncul)
modus_nilai = df['Nilai'].mode()
print(f"Nilai Modus:\n{modus_nilai}")
```
